```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras import models, layers
from sklearn.model_selection import train_test_split
```

```python
from keras.utils import image_dataset_from_directory
```

```python
train_data = image_dataset_from_directory('./afhq/train', label_mode = 'categorical', seed = 2283221337, image_size = (128,128))
valid_data = image_dataset_from_directory('./afhq/val', label_mode = 'categorical', seed = 2283221337, image_size = (128,128))
```

```python
madel = models.Sequential([
    layers.Rescaling(1./255, input_shape = (128,128,3)),
    layers.Conv2D(16,3, padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3, padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(units =128,activation = 'relu'),
    layers.Dense(units =64,activation = 'relu'),
    layers.Dense(units =3,activation = 'softmax')
])
madel.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
madel.summary()
```

```python
madel.fit(train_data, validation_data = valid_data, epochs=16, batch_size=32)
```

```python
import tensorflow as tf
from keras.utils import load_img,img_to_array
mouse = load_img('mb1wb.jpg', target_size=(128,128))
plt.imshow(mouse)
mouse = img_to_array(mouse)
mouse = tf.expand_dims(mouse, 0)
prediction = madel.predict(mouse)
score = tf.nn.softmax(prediction[0])
train_data.class_names[np.argmax(score)]
```
