```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
from keras import models, layers
from sklearn.model_selection import train_test_split
```

```python
infa = pd.read_csv('diabetes.csv')
infa
```

```python
y = infa['Outcome']
x = infa.drop(['Outcome'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
xscaled = scaler.fit_transform(x)
xtrain, xtest, ytrain, ytest =train_test_split(xscaled, y, test_size = 0.2)
```

```python
madel = models.Sequential([
    layers.Dense(units = 16, activation = 'relu', input_shape = (8, )), 
    layers.Dense(units = 16, activation = 'relu'),
    layers.Dense(units = 1, activation = 'sigmoid')
])
madel.summary()
```

```python
madel.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
madel.fit(xtrain, ytrain, epochs=64, batch_size= 32, validation_data=(xtest,ytest),verbose=2)
```
