```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
infa = pd.read_csv('diabetes.csv')
y = infa['Outcome']
x = infa.drop(['Outcome'], axis = 1)
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
xscaled = scaler.fit_transform(x)
xscaled
```

```python
from sklearn.decomposition import PCA
p = PCA(n_components=2)
pscal = p.fit_transform(xscaled)
plt.scatter(pscal[:,0], pscal[:,1], c = y)
```

```python
from sklearn.manifold import TSNE
t = TSNE(n_components=2)
tscal = t.fit_transform(xscaled)
plt.scatter(tscal[:,0], tscal[:,1], c = y)
```

```python
from sklearn.cluster import KMeans
import cv2
pict = cv2.imread('forest.jpg')
pict = cv2.cvtColor(pict,cv2.COLOR_BGR2RGB)
```

```python
pict.shape
```

```python
w,h,d = tuple(pict.shape)
```

```python
imagearray = np.reshape(pict, (w*h, d))
```

```python
from sklearn.utils import shuffle
training = shuffle(imagearray, random_state = 0, n_samples = 1000)
```

```python
def recreate(codebook, labels,w,h ):
    return codebook[labels].reshape(w,h,-1)
```

```python
plt.imshow(pict)
```

```python
kmodel64 = KMeans(n_clusters= 64, random_state = 0)
kmodel64.fit(training)
predict= kmodel64.predict(imagearray)
chtoto = recreate(kmodel64.cluster_centers_, predict, w,h) 
chtoto = chtoto.astype(int)
plt.imshow(chtoto)
```

```python
kmodel32 = KMeans(n_clusters= 32, random_state = 0)
kmodel32.fit(training)
predict= kmodel32.predict(imagearray)
chtoto = recreate(kmodel32.cluster_centers_, predict, w,h) 
chtoto = chtoto.astype(int)
plt.imshow(chtoto)
```
